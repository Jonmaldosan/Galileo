#### ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import sys

array_bool = [True] * 1000

size_in_bytes = sys.getsizeof(array_bool)

print("El tamaño en bytes del array es:", size_in_bytes)
```

#### Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np

a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

diff = a - b
mean = np.mean(diff)

print(f"La diferencia es: {diff}")
print(f"La media aritmética de la diferencia es: {mean}")
```

## Parte 2: Exploración y Minería de Datos:

```python
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
```

```python
nucpob = pd.read_csv('nucleospoblacion.csv')
nucpob
```

#### ¿Cuántos Municipios tienen más de 100000 habitantes?.

```python
municipios = (nucpob["Poblacion"] > 100000).sum()
municipios
```

####  Realice una gráfica de barras sobre la polbación de cada ciudad, ordenela de menor a mayo y responda:

```python
Pobgraph1 = nucpob[["Municipio", "Poblacion"]]
Pobgraph1 = Pobgraph1.sort_values(by='Poblacion', ascending=0)
plt.barh(Pobgraph1['Municipio'], Pobgraph1['Poblacion'])
plt.title("Poblacion por Municipio")
plt.show()
```

```python
maspob = nucpob[nucpob["Poblacion"] > 200000]
maspob
```

```python
Pobgraph2 = maspob[["Municipio", "Poblacion"]]
Pobgraph2 = Pobgraph2.sort_values(by='Poblacion', ascending=0)
plt.barh(Pobgraph2['Municipio'], Pobgraph2['Poblacion'])
plt.title("Poblacion por Municipio")
plt.show()
```
